```python
d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000],
'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'],
'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}
```

```python
df = pd.DataFrame(d)
```

```python
df.head()
```

```python
df
```

```python
# Here city is a nominal data so use One hot encoder
```

```python
df["city"].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

### Call one Hot encoder & ignore unknown values and change it in pandas df

```python
Oh_en = OneHotEncoder(handle_unknown = "ignore",sparse_output = False).set_output(transform = 'pandas')
```

```python
ohetransform = Oh_en.fit_transform(df[['city']])
```

```python
ohetransform
```

```python
df.head()
```

```python
df = pd.concat([df,ohetransform],axis = 1).drop(columns = ['city'])
```
