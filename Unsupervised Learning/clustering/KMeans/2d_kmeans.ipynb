```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student_clustering.csv')
df.head()
```

```python
plt.scatter(df['cgpa'],df['iq'])
```

```python
#Here are 4 perfectly visible clusters so i think i should be choosing k = 4
```

```python
from sklearn.cluster import KMeans
```

```python
ls =[]
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    ls.append(km.inertia_)
```

```python
ls
```

```python
plt.plot(range(1,11),ls)
plt.xlabel("No. of Clusters")
plt.ylabel("inertia")#WCSS Within Cluster Sum of Square
```

```python
# We choose that k using the elbow method so k = 4
```

```python
X = df.iloc[:,:].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```

y_means

```python
plt.scatter(X[y_means ==0,0],X[y_means == 0,1],color = 'blue')
plt.scatter(X[y_means ==1,0],X[y_means == 1,1],color = 'black')
plt.scatter(X[y_means ==2,0],X[y_means == 2,1],color = 'red')
plt.scatter(X[y_means ==3,0],X[y_means == 3,1],color = 'GREEN')
```
