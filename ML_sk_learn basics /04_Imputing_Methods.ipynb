```python
import pandas as pd
```

```python
import numpy as np
```

```python
miles = pd.DataFrame({'farthest_run_mi':[50,62,np.nan,100,26,13,31,50]})
```

```python
miles
```

```python
miles.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp_mean = SimpleImputer(strategy = 'mean')
```

```python
imp_mean.fit_transform(miles)
```

```python
imp_med = SimpleImputer(strategy = 'median')
```

```python
imp_med.fit_transform(miles)
```

```python
miles
```

```python
imp_mod = SimpleImputer(strategy = 'most_frequent')
```

```python
imp_mod.fit_transform(miles)
```

```python
imp_cons = SimpleImputer(strategy = 'constant' , fill_value = 20)
```

```python
imp_cons.fit_transform(miles)
```
